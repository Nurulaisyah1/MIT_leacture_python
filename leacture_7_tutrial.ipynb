```python
# 1. TESTING APPROACHES 


#(Natural Boundaries and Partition/Batasan dan Partisi Alami)
#pengujian sebuah fungsu is_bigger
def is_bigger(x, y):
    """ Mengembalikan True jika x lebih besar dari y, jika tidak False """
    return x > y

#pengujian
def test_natural_partition():
    #assert digunakan untuk memverifikasi bahwa hasil dari fungsi sesuai dengan yang diharapkan. Jika tidak, akan menampilkan pesan kesalahan.
    # x sama dengan y
    assert is_bigger(5, 5) == False, "Gagal: Nilai sama seharusnya False"
    
    # x lebih besar dari y
    assert is_bigger(10, 5) == True, "Gagal: x > y seharusnya True"
    
    # x lebih kecil dari y
    assert is_bigger(3, 8) == False, "Gagal: x < y seharusnya False"
    
    print("Semua pengujian partisi alami lulus.")
```

```python
#random testing / pengujian acak

import random

def test_random():
    for _ in range(5):  # lakukan 5 kali
        x = random.randint(1, 100)
        y = random.randint(1, 100)
        result = is_bigger(x, y)
        print(f"Testing is_bigger({x}, {y}) -> {result}")

test_random()
```

```python
#blackbox testing
#spesifikasi fungsi 

#def sqrt(x, eps):
# """ Assumes x, eps floats, x >= 0, eps > 0
#Returns res such that x-eps <= res*res <= x+eps """
    
# Fungsi sqrt yang sudah diimplementasikan
def sqrt(x, eps):
    if x == 0:
        return 0

    low = 0
    high = max(1, x)
    res = (low + high) / 2

    while abs(res**2 - x) > eps:
        if res**2 < x:
            low = res
        else:
            high = res
        res = (low + high) / 2

    return res

# Fungsi pengujian
def test_sqrt_black_box():
    # Kasus 1: Boundary (x = 0, eps = 0.0001)
    assert 0 - 0.0001 <= sqrt(0, 0.0001) ** 2 <= 0 + 0.0001, "Gagal: sqrt(0, 0.0001)"
    
    # Kasus 2: Perfect Square (x = 25, eps = 0.0001)
    assert 25 - 0.0001 <= sqrt(25, 0.0001) ** 2 <= 25 + 0.0001, "Gagal: sqrt(25, 0.0001)"
    
    # Kasus 3: Less than 1 (x = 0.05, eps = 0.0001)
    assert 0.05 - 0.0001 <= sqrt(0.05, 0.0001) ** 2 <= 0.05 + 0.0001, "Gagal: sqrt(0.05, 0.0001)"
    
    # Kasus 4: Irrational Square Root (x = 2, eps = 0.0001)
    assert 2 - 0.0001 <= sqrt(2, 0.0001) ** 2 <= 2 + 0.0001, "Gagal: sqrt(2, 0.0001)"
    
    # Kasus 5: Extremes (x = 2, eps = 1.0/2.0**64.0)
    assert 2 - 1.0/2.0**64.0 <= sqrt(2, 1.0/2.0**64.0) ** 2 <= 2 + 1.0/2.0**64.0, "Gagal: sqrt(2, 1.0/2.0**64.0)"
    
    # Kasus 6: Extremes (x = 1.0/2.0**64.0, eps = 1.0/2.0**64.0)
    assert 1.0/2.0**64.0 - 1.0/2.0**64.0 <= sqrt(1.0/2.0**64.0, 1.0/2.0**64.0) ** 2 <= 1.0/2.0**64.0 + 1.0/2.0**64.0, "Gagal: sqrt(1.0/2.0**64.0, 1.0/2.0**64.0)"
    
    # Kasus 7: Extremes (x = 2.0**64.0, eps = 1.0/2.0**64.0)
    assert 2.0**64.0 - 1.0/2.0**64.0 <= sqrt(2.0**64.0, 1.0/2.0**64.0) ** 2 <= 2.0**64.0 + 1.0/2.0**64.0, "Gagal: sqrt(2.0**64.0, 1.0/2.0**64.0)"
    
    # Kasus 8: Extremes (x = 1.0/2.0**64.0, eps = 2.0**64.0)
    assert 1.0/2.0**64.0 - 2.0**64.0 <= sqrt(1.0/2.0**64.0, 2.0**64.0) ** 2 <= 1.0/2.0**64.0 + 2.0**64.0, "Gagal: sqrt(1.0/2.0**64.0, 2.0**64.0)"
    
    # Kasus 9: Extremes (x = 2.0**64.0, eps = 2.0**64.0)
    assert 2.0**64.0 - 2.0**64.0 <= sqrt(2.0**64.0, 2.0**64.0) ** 2 <= 2.0**64.0 + 2.0**64.0, "Gagal: sqrt(2.0**64.0, 2.0**64.0)"
    
    print("Semua pengujian black box lulus.")

# Jalankan tes
test_sqrt_black_box()
```

```python
#GLASS BOX TESTING / WHITE BOX TESTING

def abs(x):
    """
    Mengasumsikan x adalah int
    Mengembalikan x jika x >= 0 dan -x jika sebaliknya
    """
    if x < -1:
        return -x
    else:
        return x

```

```python
#pengujian glass box
def test_abs_glass_box():
    # Kasus 1: x = -2, seharusnya mengembalikan 2
    assert abs(-2) == 2, "Gagal: abs(-2)"
    
    # Kasus 2: x = 2, seharusnya mengembalikan 2
    assert abs(2) == 2, "Gagal: abs(2)"
    
    # Kasus 3: x = -1, seharusnya mengembalikan 1 (tapi akan gagal karena bug)
    assert abs(-1) == 1, "Gagal: abs(-1), hasil yang diharapkan adalah 1"

# Jalankan tes
test_abs_glass_box()
```

```python
######################################
# EXAMPLE: Raising your own exceptions
######################################
def get_ratios(L1, L2):#membagi elemen-elemen dari dua list, L1 dan L2, yang diasumsikan memiliki panjang yang sama
    """ Assumes: L1 and L2 are lists of equal length of numbers
        Returns: a list containing L1[i]/L2[i] """
    ratios = [] #membuat daftar list kosong
    for index in range(len(L1)): #menelusuri setiap elemen berdasarkan index
        try:
            ratios.append(L1[index]/L2[index]) #index L1/L2 yang kemusian hasilnya ditambahkan kedalam list ratios.
        except ZeroDivisionError: #menambahkan float('nan') sebagai pengganti jika terjadi pembagian dengan nol.
            ratios.append(float('nan')) #nan = Not a Number
        except:
            raise ValueError('get_ratios called with bad arg')
        else:
            print("success")
        finally:
            print("executed no matter what!")
    return ratios
    
print(get_ratios([1, 4], [2, 4]))
```

```python
#######################################
## EXAMPLE: Exceptions and lists
#######################################
def get_stats(class_list): #class_list adalah list yang berisi data siswa
	new_stats = [] #membuat daftar list baru 
	for person in class_list: 
		new_stats.append([person[0], person[1], avg(person[1])]) #avg(person[1]) : fungsi avg adalah menghitung rata-rata
	return new_stats #mengembalikan nilai kedalam list ner_stats

# avg function: version without an exception
#def avg(grades):
#    return (sum(grades))/len(grades)
    
# avg function: version with an exception
def avg(grades): #menghitung rata-rata tanpa memeriksa apakah daftar nilai kosong
    try:
        return sum(grades)/len(grades) #Menghitung jumlah nilai dengan sum(grades), lalu membagi dengan jumlah elemen dalam list len(grades)
    except ZeroDivisionError:#Jika daftar grades kosong, ini akan memicu ZeroDivisionError karena kamu membagi dengan nol.
        print('warning: no grades data')
        return 0.0


# avg function: version with assert
def avg(grades): #Fungsi ini menggunakan assertion untuk memastikan bahwa daftar nilai tidak kosong sebelum melakukan pembagian.
    assert len(grades) != 0, 'warning: no grades data' #assert memastikan bahwa panjang daftar nilai len(grades) tidak boleh nol.
    return sum(grades)/len(grades)

    
test_grades = [[['peter', 'parker'], [80.0, 70.0, 85.0]], 
              [['bruce', 'wayne'], [100.0, 80.0, 74.0]],
              [['captain', 'america'], [80.0, 70.0, 96.0]],
              [['deadpool'], []]]

print(get_stats(test_grades))
```
